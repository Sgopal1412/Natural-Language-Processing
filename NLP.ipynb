```python
#Using spacy, separate the file into sentences and tokens.  Count the number of  unique tokens and sentences in the text.
import spacy
nlp = spacy.load("en_core_web_sm")
with open("/content/drive/MyDrive/nlpdoc.txt", "r", encoding="utf-8") as file:
    text = file.read()
print(text)
doc = nlp(text)

tokens = []
sentences = []
for sentence in doc.sents:
    sentences.append(sentence.text)
for token in doc:
    tokens.append(token.text)

unique_tokens = set(tokens)
num_unique_tokens = len(unique_tokens)
num_sentences = len(sentences)
print("\nNumber of unique tokens:", num_unique_tokens)
print("Number of sentences:", num_sentences)
```

```python
import os
from moviepy.editor import VideoFileClip, concatenate_videoclips

def get_video_files_from_folder(folder_path):
    video_files = []
    for file in os.listdir(folder_path):
        if file.endswith(('.mp4', '.avi', '.mov', '.mkv')):
            video_files.append(os.path.join(folder_path, file))
    return sorted(video_files)

# Folder containing the video files
input_folder_path = '/home/gopalakrishnan/Downloads/www.1TamilMV.cz - Brinda (2024) S01 EP (01-08) - 1080p - HQ HDRip - [Tam + Tel + Hin + Mal + Kan] - AAC - 8GB - ESub'

# Folder to save the output video
output_folder_path = '/home/gopalakrishnan/Downloads/outputvideo'

# Ensure the output folder exists
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Construct the output file path
output_file_name = 'outputvideo.mp4'
output_file_path = os.path.join(output_folder_path, output_file_name)

# Get the list of video files
video_files = get_video_files_from_folder(input_folder_path)

if not video_files:
    print("No video files found in the specified folder.")
else:
    # Load all video clips
    video_clips = []
    for video in video_files:
        try:
            clip = VideoFileClip(video)
            if clip.duration:  # Check if the clip has a valid duration
                video_clips.append(clip)
            else:
                print(f"Skipping {video} because it has no valid duration.")
        except Exception as e:
            print(f"Error loading {video}: {e}")

    if video_clips:
        try:
            # Concatenate video clips
            final_clip = concatenate_videoclips(video_clips, method="compose")

            # Write the final video to a file
            final_clip.write_videofile(output_file_path, codec='libx264')

            print(f"Video saved to {output_file_path}")
        except Exception as e:
            print(f"Error concatenating or writing video: {e}")
    else:
        print("No valid video clips to concatenate.")
```

```python
import spacy
from collections import Counter


nlp = spacy.load("en_core_web_sm")


text='''“Have you never thought of marrying your prodigal son Anatole?” she
asked. “They say old maids have a mania for matchmaking, and though I
don’t feel that weakness in myself as yet, I know a little person who
is very unhappy with her father. She is a relation of yours, Princess
Mary Bolkónskaya.”

Prince Vasíli did not reply, though, with the quickness of memory and
perception befitting a man of the world, he indicated by a movement of
the head that he was considering this information.

“Do you know,” he said at last, evidently unable to check the sad
current of his thoughts, “that Anatole is costing me forty thousand
rubles a year? And,” he went on after a pause, “what will it be in
five years, if he goes on like this?” Presently he added: “That’s
what we fathers have to put up with.... Is this princess of yours
rich?”

“Her father is very rich and stingy. He lives in the country. He is
the well-known Prince Bolkónski who had to retire from the army under
the late Emperor, and was nicknamed ‘the King of Prussia.’ He is
very clever but eccentric, and a bore. The poor girl is very unhappy.
She has a brother; I think you know him, he married Lise Meinen lately.
He is an aide-de-camp of Kutúzov’s and will be here tonight.”

“Listen, dear Annette,” said the prince, suddenly taking Anna
Pávlovna’s hand and for some reason drawing it downwards. “Arrange
that affair for me and I shall always be your most devoted slave-slafe
with an f, as a village elder of mine writes in his reports. She is rich
and of good family and that’s all I want.”

And with the familiarity and easy grace peculiar to him, he raised the
maid of honor’s hand to his lips, kissed it, and swung it to and fro
as he lay back in his armchair, looking in another direction.

“Attendez, said Anna Pávlovna, reflecting, “I’ll speak to
Lise, young Bolkónski’s wife, this very evening, and perhaps the
thing can be arranged. It shall be on your family’s behalf that I’ll
start my apprenticeship as old maid.”

'''

doc = nlp(text)

def extract_bigrams(doc):
    bigrams = []
    for token in doc[:-1]:
        if not token.is_punct and not doc[token.i + 1].is_punct:
            bigrams.append((token.text, doc[token.i + 1].text))
    return bigrams


def extract_trigrams(doc):
    trigrams = []
    for token in doc[:-2]:
        if not token.is_punct and not doc[token.i + 1].is_punct and not doc[token.i + 2].is_punct:
            trigrams.append((token.text, doc[token.i + 1].text, doc[token.i + 2].text))
    return trigrams
def extract_noun_phrases(doc):
    noun_phrases = [chunk.text for chunk in doc.noun_chunks]
    return noun_phrases
def count_word_frequency(doc):
    word_freq = Counter([token.text for token in doc if not token.is_punct and not token.is_space])
    return word_freq
bigrams = extract_bigrams(doc)
trigrams = extract_trigrams(doc)
noun_phrases = extract_noun_phrases(doc)
word_freq = count_word_frequency(doc)
print("All Bigrams:")
print(bigrams)
print("\nAll Trigrams:")
print(trigrams)
print("\nAll Noun Phrases:")
print(noun_phrases)
print("\nTop 10 Most Frequent Words:")
print(word_freq.most_common(10))
```

```python
import re
import spacy
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Load the text from the novel
text='''“Have you never thought of marrying your prodigal son Anatole?” she
asked. “They say old maids have a mania for matchmaking, and though I
don’t feel that weakness in myself as yet, I know a little person who
is very unhappy with her father. She is a relation of yours, Princess
Mary Bolkónskaya.”

Prince Vasíli did not reply, though, with the quickness of memory and
perception befitting a man of the world, he indicated by a movement of
the head that he was considering this information.

“Do you know,” he said at last, evidently unable to check the sad
current of his thoughts, “that Anatole is costing me forty thousand
rubles a year? And,” he went on after a pause, “what will it be in
five years, if he goes on like this?” Presently he added: “That’s
what we fathers have to put up with.... Is this princess of yours
rich?”

“Her father is very rich and stingy. He lives in the country. He is
the well-known Prince Bolkónski who had to retire from the army under
the late Emperor, and was nicknamed ‘the King of Prussia.’ He is
very clever but eccentric, and a bore. The poor girl is very unhappy.
She has a brother; I think you know him, he married Lise Meinen lately.
He is an aide-de-camp of Kutúzov’s and will be here tonight.”

“Listen, dear Annette,” said the prince, suddenly taking Anna
Pávlovna’s hand and for some reason drawing it downwards. “Arrange
that affair for me and I shall always be your most devoted slave-slafe
with an f, as a village elder of mine writes in his reports. She is rich
and of good family and that’s all I want.”

And with the familiarity and easy grace peculiar to him, he raised the
maid of honor’s hand to his lips, kissed it, and swung it to and fro
as he lay back in his armchair, looking in another direction.

“Attendez, said Anna Pávlovna, reflecting, “I’ll speak to
Lise, young Bolkónski’s wife, this very evening, and perhaps the
thing can be arranged. It shall be on your family’s behalf that I’ll
start my apprenticeship as old maid.”

india,southafrica'''

# Extract named entities (persons and locations) using spaCy
doc = nlp(text)

# Filter and count person names and locations
characters = Counter()
locations = Counter()

for ent in doc.ents:
    if ent.label_ == "PERSON":
        # Split the character name into individual words and consider only the first word
        character_name = ent.text.split()[0]
        characters[character_name] += 1
    elif ent.label_ == "GPE" or ent.label_ == "LOC":  # Change the condition to include GPE or LOC
        locations[ent.text] += 1

# Generate word clouds for Characters
wordcloud_char = WordCloud(width=400, height=400, background_color="white").generate_from_frequencies(characters)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_char, interpolation="bilinear")
plt.axis("off")
plt.title("Character Word Cloud")
plt.show()

wordcloud_loc = WordCloud(width=400, height=100, background_color="white").generate_from_frequencies(locations)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_loc, interpolation="bilinear")
plt.axis("off")
plt.title("Location Word Cloud")
plt.show()
```
